```python
import pandas as pd
import joblib

# Load model
model = joblib.load("model.pkl")

# Create empty row with ALL features
new_data = pd.DataFrame([{
    'Age': 30,
    'Income': 60000,
    'LoanAmount': 20000,
    'CreditScore': 650,
    'YearsExperience': 5,

    'Gender_Male': 1,

    'Education_High School': 0,
    'Education_Masters': 0,
    'Education_PhD': 0,

    'City_Houston': 0,
    'City_New York': 1,
    'City_San Francisco': 0,

    'EmploymentType_Self-Employed': 0,
    'EmploymentType_Unemployed': 0
}])

# Predict
prediction = model.predict(new_data)

# Output
if prediction[0] == 1:
    print("Loan Approved")
else:
    print("Loan Not Approved")
```
